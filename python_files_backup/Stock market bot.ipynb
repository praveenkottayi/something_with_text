# Import packages 

```python
import IPython
from gtts import gTTS
import os
from nsetools import Nse
import speech_recognition as sr
nse = Nse()
#print(nse)
```

# Hello Sir/madam , Please Speak to me ....  

```python
tts = gTTS(text= "Hello Sir , Please tell , about which stock you want to know", lang='en')
tts.save("welcome.mp3")
os.system("mpg321 welcome.mp3")
IPython.display.Audio("welcome.mp3")
```

```python
# obtain audio from the microphone
r = sr.Recognizer()
with sr.Microphone() as source:
    print("Energy threshold is "+ str(r.energy_threshold))
    r.adjust_for_ambient_noise(source)  # listen for 1 second to calibrate the energy threshold for ambient noise levels
    print("Hello Sir , Please tell , about which stock you want to know :) ")
    audio = r.listen(source)
    
# recognize speech using Google Speech Recognition
try:
    # for testing purposes, we're just using the default API key
    # to use another API key, use `r.recognize_google(audio, key="GOOGLE_SPEECH_RECOGNITION_API_KEY")`
    # instead of `r.recognize_google(audio)`
    print("Google Speech Recognition thinks you said : " + r.recognize_google(audio))
except sr.UnknownValueError:
    print("Google Speech Recognition could not understand audio")
except sr.RequestError as e:
    print("Could not request results from Google Speech Recognition service; {0}".format(e))
```

```python
Quote_of = r.recognize_google(audio)

if(nse.is_valid_code(Quote_of)):
    q = nse.get_quote(Quote_of) # it's ok to use both upper or lower case for codes.
    from pprint import pprint # just for neatness of display
    #pprint(q)

    stock_value = q['buyPrice1']
    tell=("Current stock value is : " +str(stock_value))
    print(tell)

    tts = gTTS(text= str(tell), lang='en')
    tts.save("price.mp3")
    os.system("mpg321 price.mp3")
    IPython.display.Audio("price.mp3")
else :
    print("I didn't find any stock with that name. Please say it again.")
```
